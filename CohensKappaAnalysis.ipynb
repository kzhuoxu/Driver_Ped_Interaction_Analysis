```python
#open "driving survey" csv file, it doesn't have a header
df = pd.read_csv('scenario3.csv')
df
```

```python
#now, go row by row, and get the values for the following: structure: A-301. A (label), 3 (scenario), 01 (question number), 301 (question). these are in column 7
labels = []
scenarios = []
questions = []
question_numbers = []

for index, row in df.iterrows():
    #get the value in column 7
    value = row.iloc[7]
    #split the value
    split_value = value.split('-')
    #get the label
    label = split_value[0]
    #get the scenario
    #print(split_value)
    scenario = split_value[1][0]
    #get the question number
    question_number = split_value[1][1:3]
    #get the question
    question = split_value[1]
    #append to lists
    labels.append(label)
    scenarios.append(scenario)
    questions.append(question)
    question_numbers.append(question_number)
#print(labels, scenarios, question_numbers, questions)

#start new df wtih these lists and column 11
new_df = pd.DataFrame({'label': labels, 'scenario': scenarios, 'question_number': question_numbers, 'question': questions, 'response': df.iloc[:,11]})
new_df
```
